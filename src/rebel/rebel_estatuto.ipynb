```python
import os 

def cargar_txt_desde_directorio(directorio):
    archivos_txt = [f for f in os.listdir(directorio) if f.endswith('.txt')]
    contenido_archivos = {}

    for archivo in archivos_txt:
        ruta_archivo = os.path.join(directorio, archivo)
        with open(ruta_archivo, 'r', encoding='utf-8') as f:
            contenido_archivos[archivo] = f.read()
    
    return contenido_archivos

# Ejemplo de uso
directorio = bas+"/Joel_Pardo/TeresIA/estatuto_goldstandard/data/gold_standard/articles"  # Reemplaza con el path a tu directorio
contenidos = cargar_txt_desde_directorio(directorio)

print(len(contenidos))

# Imprime los nombres de los archivos y sus contenidos
#for nombre_archivo, contenido in contenidos.items():
#    print(f"Contenido de {nombre_archivo}:\n{contenido}\n")
```

```python
import re

def eliminar_patron(texto):
    # Definir el patrón de la expresión regular
    patron = r'Artículo \d+\. '
    
    # Usar re.sub() para eliminar el patrón del texto
    resultado = re.sub(patron, '', texto)
    
    return resultado
```

## Pipeline usage

```python
# Importar la función pipeline de la biblioteca transformers, que facilita la inferencia con modelos preentrenados
from transformers import pipeline
from tqdm.notebook import tqdm

# Crear un pipeline de traducción utilizando un modelo y tokenizador específicos. Este modelo parece estar entrenado para extraer tripletas de relaciones.
triplet_extractor = pipeline('translation_xx_to_yy', model='Babelscape/mrebel-large', tokenizer='Babelscape/mrebel-large')

resultados_finales = []

# Solicitar al usuario que introduzca un texto para la extracción de tripletas
for i in tqdm(values3):
    result = [] 
    
    for inputt in tqdm(i):
    
        # Tokenizar manualmente el texto introducido, añadiendo tokens especiales necesarios para el modelo y obteniendo los ID de tokens generados
        tokenized = [triplet_extractor(inputt, decoder_start_token_id=250058, src_lang="es_XX", tgt_lang="<triplet>", return_tensors=True, return_text=False)[0]["translation_token_ids"]]
        
        # Decodificar los ID de tokens para obtener el texto generado por el modelo
        extracted_text = triplet_extractor.tokenizer.batch_decode(tokenized)
        
        
        # Imprimir el texto extraído
        for i in extracted_text[0].split("<triplet>"):
            if i != "<s>" or i != "</s>":
                print(i+"\n")
                
        print("-"*50)
        
        
        
        
        # Definir una función para analizar el texto generado y extraer las tripletas, incluyendo los tipos de sujeto y objeto
        def extract_triplets_typed(text):
            triplets = []  # Lista para almacenar las tripletas extraídas
            relation = ''  # Variable para almacenar la relación
            text = text.strip()  # Limpiar espacios al inicio y al final del texto
            current = 'x'  # Variable para rastrear el componente actual de la tripleta
            subject, relation, object_, object_type, subject_type = '','','','',''  # Inicializar variables para cada componente de la tripleta y sus tipos
        
            # Iterar sobre los tokens del texto, excluyendo tokens especiales y limpiando ciertos tokens adicionales
            for token in text.replace("<s>", "").replace("<pad>", "").replace("</s>", "").replace("tp_XX", "").replace("__en__", "").split():
                if token == "<triplet>" or token == "<relation>":
                    # Cambiar el componente actual a 't' (tripleta) o 'relation' (relación)
                    current = 't'
                    # Si ya hay una relación, añadir la tripleta a la lista y reiniciar las variables
                    if relation != '':
                        triplets.append({'head': subject.strip(), 'head_type': subject_type, 'type': relation.strip(),'tail': object_.strip(), 'tail_type': object_type})
                        relation = ''
                    subject = ''
                elif token.startswith("<") and token.endswith(">"):
                    # Si el token es un marcador de tipo, actualizar el tipo de sujeto u objeto y cambiar el componente actual
                    if current == 't' or current == 'o':
                        current = 's'
                        if relation != '':
                            triplets.append({'head': subject.strip(), 'head_type': subject_type, 'type': relation.strip(),'tail': object_.strip(), 'tail_type': object_type})
                        object_ = ''
                        subject_type = token[1:-1]  # Extraer el tipo del sujeto
                    else:
                        current = 'o'
                        object_type = token[1:-1]  # Extraer el tipo del objeto
                        relation = ''
                else:
                    # Añadir el token al componente actual de la tripleta
                    if current == 't':
                        subject += ' ' + token
                    elif current == 's':
                        object_ += ' ' + token
                    elif current == 'o':
                        relation += ' ' + token
            # Añadir la última tripleta si todos los componentes están presentes
            if subject != '' and relation != '' and object_ != '' and object_type != '' and subject_type != '':
                triplets.append({'head': subject.strip(), 'head_type': subject_type, 'type': relation.strip(),'tail': object_.strip(), 'tail_type': object_type})
            return triplets
        
        # Utilizar la función definida para extraer las tripletas del texto generado
        extracted_triplets = extract_triplets_typed(extracted_text[0])
        
        # Imprimir las tripletas extraídas
        for i in extracted_triplets:
            print(str(i)+"\n")
            result.append(str(i))
            
    resultados_finales.append(result)

    print("-"*100)
```

```python
import pandas as pd

def convertir_a_brat(data_df, contenidos):
    brat_annotations = {}
    textos = {}
    # Procesar cada artículo en el DataFrame
    for article in data_df['article'].unique():
        texto = " ".join(contenidos[article])  # Obtener el texto correspondiente al artículo
        df_article = data_df[data_df['article'] == article]  # Filtrar las filas del DataFrame por artículo
        
        entities = []  # Lista para almacenar entidades anotadas
        relations = []  # Lista para almacenar relaciones anotadas
        entity_counter = 1  # Contador de entidades para asignar T1, T2, etc.
        relation_counter = 1  # Contador de relaciones para asignar R1, R2, etc.

        # Crear anotaciones para entidades y relaciones
        for idx, row in df_article.iterrows():
            # Obtener las posiciones de 'head' y 'tail' en el texto
            head_pos = texto.find(row['head'])
            tail_pos = texto.find(row['tail'])

            if head_pos != -1 and tail_pos != -1:
                # Calcular los offsets para 'head'
                head_start = head_pos
                head_end = head_start + len(row['head'])

                # Calcular los offsets para 'tail'
                tail_start = tail_pos
                tail_end = tail_start + len(row['tail'])

                # Anotar la entidad 'head'
                entities.append(f"T{entity_counter}\t{row['head_type']} {head_start} {head_end}\t{row['head'].strip()}")
                head_entity = f"T{entity_counter}"
                entity_counter += 1

                # Anotar la entidad 'tail'
                entities.append(f"T{entity_counter}\t{row['tail_type']} {tail_start} {tail_end}\t{row['tail'].strip()}")
                tail_entity = f"T{entity_counter}"
                entity_counter += 1

                # Anotar la relación entre 'head' y 'tail'
                relations.append(f"R{relation_counter}\t{row['type']} Arg1:{head_entity} Arg2:{tail_entity}")
                relation_counter += 1

        # Guardar las anotaciones para este artículo
        brat_annotations[article] = '\n'.join(entities + relations)
        textos[article] = texto

    return textos, brat_annotations

# Ejemplo de uso
# Suponiendo que `data_df` es el DataFrame con las columnas ['article', 'head', 'head_type', 'type', 'tail', 'tail_type']
# Y que `contenidos` es un diccionario con los textos de cada artículo
textos, brat_annotations = convertir_a_brat(data_df, contenidos)

# Guardar los archivos en formato .ann
for article, annotation in brat_annotations.items():
    with open(f'{bas}/Joel_Pardo/TeresIA/estatuto_rebel/{article.replace(".txt", "")}.ann', 'w', encoding='utf-8') as f:
        f.write(annotation)

for article, texto in textos.items():
    with open(f'{bas}/Joel_Pardo/TeresIA/estatuto_rebel/{article}', 'w', encoding='utf-8') as f:
        f.write(texto)
```
